****OUTLIERS****

```python
# extreme high value or extreme low value are called as outliers
# outliers can be done for NUMERICAL columns
# it has 2 methods
# trimming -- remove the outliers
# replace --- replace the outliers with low value or high value in IQR range
```

```python
df["seatcapacity"].plot.box(vert=False)
```

```python
q1=df["seatcapacity"].quantile(0.25)
print("q1:",q1)
q3=df["seatcapacity"].quantile(0.75)
print("q3:",q3)
```

```python
iqr=q3-q1
print("iqr:",iqr)
```

```python
lowest_value= q1-1.5*(iqr)
print("lowest_value:",lowest_value)
highest_value= q3+1.5*(iqr)
print("highest_value:",highest_value)
```

```python
k=df[(df["seatcapacity"]>=lowest_value) & (df["seatcapacity"]<=highest_value)]
```

```python
k.shape # after trimming the outliers
```

```python
df.shape # before trimming the outliers
```

```python
1474762-1177525 # 29,700 rows

# after removing the outliers, i have stored the values in variable "K" ("k" does not have outliers)
```

```python
k["seatcapacity"].plot.box(vert=False)
```

```python
def replace(x):
    if x<lowest_value:
       x=lowest_value
    elif x>highest_value:
         x=highest_value
    else:
        x
    return x
```

```python
replace(-0.000001)
```

```python
replace(100)
```

```python
df["seatcapacity"]=df["seatcapacity"].apply(replace)
```

```python
df["seatcapacity"].plot.box(vert=False) # after replacing the outliers
```
